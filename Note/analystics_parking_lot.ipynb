```python
df = pd.read_csv("/Users/restitutor/Documents/Restitutor/Workspace/Flutter-Python/Python/Team_PL/Data/한강공원 주차장 일별 이용 현황.csv", encoding="cp949")
df[df["지구별"] == "PLT-008"]
```

```python
df["year"] = pd.to_datetime(df["날짜"]).dt.year
df["month"] = pd.to_datetime(df["날짜"]).dt.month
df["day"] = pd.to_datetime(df["날짜"]).dt.day
df.drop(
    columns=["날짜","이용시간"],
    inplace=True
)
df.head()
```

```python
import holidays

kr_holidays = holidays.KR(years=[2023, 2024, 2025])
df["날짜"] = pd.to_datetime(df[["year", "month", "day"]])
df["날짜 타입"] = df["날짜"].apply(
    lambda x: "휴일" if (x.weekday() >= 5 or x.date() in kr_holidays) else "평일"
)

sp = df[(df["지구별"] == "PLT-008") & (df["year"] >= 2023) & (df["year"] < 2026)]
sp.drop(
    columns=["지구별", "month", "day", "날짜"],
    inplace=True
)
```

```python
tYear = [2023, 2024, 2025]
for y in tYear:\
sp[sp["year"] == y].reset_index(
    drop=True,
).to_csv(f"../Data/Target_{y}_raw.csv", index=False)
```
